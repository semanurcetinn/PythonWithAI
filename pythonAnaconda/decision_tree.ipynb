```python
# belirleyici durumlar arar, NOX > 50 --> NOX < 50
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("HousingDataset_Clean.csv", index_col=0)
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=["MEDV"]) #etiketinin kaldırılmış hali
y = df["MEDV"] #sadece etiketli hali
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth = 4, random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#verinin ortalamasından sapan değer
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
r2 #r2 1 olduğunda genellikle overfitting olabilir
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names = x.columns, filled = True, rounded = True)
plt.title("Decision Tree")
```

```python
import seaborn as sb

plt.figure(figsize=(10,6))
sb.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor(n_estimators = 10, random_state = 42) #kaç tane decision tree istiyorsak n_estimators kısmında belirtiriz.
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
mae2 = mean_absolute_error(y_test, y_pred2)
mae2
```

```python
# GridSearchCV, RandomSearchCV -> en iyi parametlereleri bulma yöntemleri
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": list(range(5,20))
}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid, scoring="r2")
grid.fit(X_train, y_train)
```

```python
grid.best_params_, grid.best_score_
```
