```python
import pandas as pd
```

```python
df= pd.read_csv("possum.csv")
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["age"].fillna(df["age"].mean(), inplace= True)
df["footlgth"].fillna(df["footlgth"].mean(), inplace= True)
df
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=["age", "Pop", "sex"])
y = df["age"]
x
```

```python
#linear regression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.9, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_linear = model.predict(X_test)
y_pred_linear
```

```python
import matplotlib.pyplot as plot
```

```python
plot.scatter(y_test, y_pred_linear, label="Tahminler")
plot.xlabel("Gerçek Yaşlar")
plot.ylabel("Tahmin Edilen Yaşlar")
plot.legend()

min_val = min(y_test.min(), y_pred_linear.min())
max_val = max(y_test.max(), y_pred_linear.max())

plot.plot([min_val,max_val], [min_val, max_val], color= "red", linestyle="--")

plot.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred_linear)
mse = mean_squared_error(y_test, y_pred_linear)
r2 = r2_score(y_test, y_pred_linear)

mae, mse, r2
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
```

```python
model2 = DecisionTreeRegressor(max_depth=5, random_state=42)
model2.fit(X_train, y_train)
```

```python
y_pred_decision_tree = model2.predict(X_test)
```

```python
from sklearn.tree import plot_tree
plot.figure(figsize=(20,10))
plot_tree(model2, feature_names = x.columns, filled = True, rounded = True)
plot.title("Decision Tree")
```

```python
model2_mae = mean_absolute_error(y_test, y_pred_decision_tree)
model2_mse = mean_squared_error(y_test, y_pred_decision_tree)
model2_r2 = r2_score(y_test, y_pred_decision_tree)

model2_mae, model2_mse, model2_r2
```
