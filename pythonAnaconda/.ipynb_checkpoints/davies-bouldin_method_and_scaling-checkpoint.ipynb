```python
# Davies-Bouldin Method
```

Kümelerin birbirinden ne kadar iyi ayrıldığını ve kendi içinde ne kadar yoğun olduğunu ölçen bir sayıdır. Skor ne kadar küçükse, kümelenme o kadar iyidir. Kümeler arası mesafe büyük, kümeler içi sıkı anlamına gelir.

Küme içindeki elemanların, o kümenin merkezine olan ortalama uzaklığına bakılır.

Sonra bu kümeyle diğer tüm kümeler karşılaştırılır.

İki küme arasındaki merkezler arası uzaklık bulunur.

Ardından, iki kümenin yoğunlukları toplanır ve merkezler arası uzaklığa bölünür.

Tüm kümeler için bu değerlerin ortalaması alınır

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)
```

```python
k_means = KMeans(n_clusters=4, random_state=42)
labels = k_means.fit_predict(X)
```

```python
db_index = davies_bouldin_score(X, labels)
print("Davis-Bouldin Skoru:", db_index)
#0'a ne kadar yakınsa, kümeler o kadar başarılı ayrılmış ve yoğun demektir
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], c='red', marker='*', s=200)
plt.title("KMeans Kümeleme")
plt.show()
```

```python
# Scaling (ölçekleme)
```

Makine öğrenmesinde veri değerlerini belli bir aralığa getirme işlemidir. Farklı ölçeklerdeki sayılarla çalışırken değişkenlerin büyüklüğünden etkilenmemek için verileri eşit düzeye çekerek algoritmaların daha adil ve doğru çalışmasını sağlar.
Veri karar ağaçları, kurallara dayanmayıp mesafeye, doğrusal ilişkiye dayanıyorsa Scaling zorunludur. 

Min-Max Scaling -> Tüm verileri 0 ile 1 arasına getirir. Sinir ağları

Standard Scaling (Z-score normalizasyonu) -> Veriyi ortalaması 0, standart sapması 1 olacak şekilde dönüştürür. KNN, PCA, SVM

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np
```

```python
X = np.array([[1,100],[2,200],[3,300]])

minmax = MinMaxScaler()
X_minmax = minmax.fit_transform(X)

standard = StandardScaler()
X_standard = standard.fit_transform(X)

print("Min-Max:", X_minmax)
print("Standard:", X_standard)
```
