```python
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)
#veri üretme sayısı 300, verilerin grup sayısı 4
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y_true, s=50) #s->size noktaların büyüklüğü, c-> renkleri
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
X = df[ ["Annual Income (k$)", "Spending Score (1-100)"] ].values
X
```

```python
from sklearn.cluster import KMeans
#belirlediğimiz sayıda rastgele merkezler belirler, 
#her veri noktası kendisini en yakın merkez noktaya atar
#sonra her küme için yeni bir merkez hesaplar
model = KMeans(n_clusters=5)
model.fit(X)
```

```python
model.cluster_centers_ #bulduğu grupları gösterir, merkez noktalarındaki verileri
```

```python
y = model.predict(X) #her veriyi belirlediği grup merkezine göre en yakın gruba atadı
y
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y) #x'in bütün 0. indexleri, y'nin bütün 1. indexleri
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=200, marker="*", c="red")
plt.xlabel("yıllık gelir (k$)")
plt.ylabel("harcama skoru (1-100)")
plt.show()
```

```python
model.inertia_
```

```python
#inertia -> her bir noktaın kendi küme merkezine olan uzaklığının karelerinin toplamıdır.
#noktalar veri merkezine ne kadar yakınsa inertia o kadar düşüktür ve düşük inertia skoru daha iyidir
#küme içi tutarlılık için küme dışıyla ilgilenmiyorsam elbow method
def elbow_method():
    inertias = [] #belirli kmeans algoritmaları eğitip en az inertia skoruna sahip olan en iyi skordur 
    for k in range(2,11): # 2 gruptan başlayıp 10 gruba kadar kmeans algoritması geliştir
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)
    plt.figure(figsize=(8,6))
    plt.plot(range(2,11), inertias)

elbow_method()
#dirsek noktası en iyi noktadır.
```

```python
# silhouette method -> kümenin elemanlarını küme noktasına olan uzaklığına bakar
#kendi kümesine yakınlık + diğer kümelere uzaklık
#kohezyon seperation küme içi tutarlılık ve küme dışı ayrışmanın aynı anda iyi olmasını istiyorsam sihouette method
from sklearn.metrics import silhouette_score

def my_silhouette_method():
    scores = []
    for k in range(2,11):
        kmeans = KMeans(n_clusters=k,random_state=42)
        y = kmeans.fit_predict(X) #eğitip anında tahmin ettirir
        scores.append(silhouette_score(X,y))
    plt.figure(figsize=(8,6))
    plt.plot(range(2,11), scores)
    plt.show()
my_silhouette_method()
#1'e yaklaştıkça başarı oranı artar
```
